```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import hamming_loss
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import make_pipeline
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.spatial import distance
import random
import warnings
import operator
import statistics
warnings.filterwarnings('ignore')
```

### 1 (a)

```python
data = pd.read_csv("../data/Frogs_MFCCs.csv")
Xdata = data.iloc[:,:-4]
Ydata = data.iloc[:,-4:]
YDataFamily = Ydata['Family']
YDataGenus = Ydata['Genus']
YDataSpecies = Ydata['Species']
XTrain, XTest, YTrain, YTest = train_test_split(Xdata, Ydata, test_size=0.3)
```

### 1 (b) - i

### Exact match: It ignore partially  correct, as it consider them incorrectand further extend accuracy used in single label case for multi-label prediction.Disadvantage of using this method is that it does not distinguish between complete incorrect and partially correct. 

### Hamming loss: It gives us average of the relevance of an example to a class label is incorrectly predicted. Further hamming loss consider prediction error and missing error, normalised over total number of classes and total number of examples. Ideal condition will be having zero hamming loss which states no error. Therefore smaller value of hamming loss gives better performance of an algorithm 

### 1 (b) - ii

```python
YTrainFamily = YTrain['Family']
YTrainGenus = YTrain['Genus']
YTrainSpecies = YTrain['Species']

YTestFamily = YTest['Family']
YTestGenus = YTest['Genus']
YTestSpecies = YTest['Species']
```

```python
familySVC = SVC(kernel= "rbf", decision_function_shape = "ovr")
familyModel = GridSearchCV(familySVC,{"C":[0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000, 100000, 1000000], "gamma":[0.01, 0.04, 0.06, 0.1, 0.3, 1 , 1.5, 1.9, 2.1, 2.3, 2.7]}, cv=10).fit(XTrain, YTrainFamily)

familyModelC = familyModel.best_params_['C']
familyModelgamma = familyModel.best_params_['gamma']
print('Family C value:',familyModelC)
print('Family gamma value:',familyModelgamma)

familySVCTest = SVC(C =familyModelC, gamma=familyModelgamma, kernel= "rbf", decision_function_shape = "ovr").fit(XTrain, YTrainFamily)
prediction = familySVCTest.predict(XTest)
familyhammingLoss = hamming_loss(YTestFamily, prediction)
print('Family label Hamming Loss:', familyhammingLoss)
print('Family labe Hamming Score:', 1 - familyhammingLoss)
```

```python
genusSVC = SVC(kernel= "rbf", decision_function_shape = "ovr")
genusModel = GridSearchCV(genusSVC,{"C":[0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000, 100000, 1000000], "gamma":[0.01, 0.04, 0.06, 0.1, 0.3, 1 , 1.5, 1.9, 2.1, 2.3, 2.7]}, cv=10).fit(XTrain, YTrainGenus)

genusModelC = genusModel.best_params_['C']
genusModelgamma = genusModel.best_params_['gamma']
print('Genus C value:',genusModelC)
print('Genus gamma value:',genusModelgamma)

genusSVCTest = SVC(C =genusModelC, gamma=genusModelgamma, kernel= "rbf", decision_function_shape = "ovr").fit(XTrain, YTrainGenus)
prediction = genusSVCTest.predict(XTest)
genushammingLoss = hamming_loss(YTestGenus, prediction)
print('Genus label Hamming Loss:', genushammingLoss)
print('Genus labe Hamming Score:', 1 - genushammingLoss)
```

```python
speciesSVC = SVC(kernel= "rbf", decision_function_shape = "ovr")
speciesModel = GridSearchCV(speciesSVC,{"C":[0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000, 100000, 1000000], "gamma":[0.01, 0.04, 0.06, 0.1, 0.3, 1 , 1.5, 1.9, 2.1, 2.3, 2.7]}, cv=10).fit(XTrain, YTrainSpecies)

speciesModelC = speciesModel.best_params_['C']
speciesModelgamma = speciesModel.best_params_['gamma']
print('Species C value:',speciesModelC)
print('Species gamma value:',speciesModelgamma)

speciesSVCTest = SVC(C =speciesModelC, gamma=speciesModelgamma, kernel= "rbf", decision_function_shape = "ovr").fit(XTrain, YTrainSpecies)
prediction = speciesSVCTest.predict(XTest)
specieshammingLoss = hamming_loss(YTestSpecies, prediction)
print('Species label Hamming Loss:', specieshammingLoss)
print('Species labe Hamming Score:', 1 - specieshammingLoss)
```

### 2 (a)

```python
bestK = 0
bestScore = 0
bestlabels = []
for i in range(2,51):
    model = KMeans(n_clusters=i).fit(Xdata)
    feat = model.labels_
    score = silhouette_score(Xdata, feat)
    if score > bestScore:
        bestScore = score
        bestK = i
        bestlabels = feat
print('best k value:', bestK)
print('best Silhouettes score:', bestScore)
```

### 2 (b)

```python
from collections import defaultdict

getClusterData = defaultdict(list)
bestlabels = pd.Series(bestlabels)
for i in range(bestK):
    getFamilyCluster = dict(YDataFamily[bestlabels.index[bestlabels==i].tolist()].value_counts())
    getClusterData[i].append((next(iter((getFamilyCluster.items())))[0],next(iter((getFamilyCluster.items())))[1]))
    
    getGenusCluster = dict(YDataGenus[bestlabels.index[bestlabels==i].tolist()].value_counts())
    getClusterData[i].append((next(iter((getGenusCluster.items())))[0],next(iter((getGenusCluster.items())))[1]))
    
    getSpeciesCluster = dict(YDataSpecies[bestlabels.index[bestlabels==i].tolist()].value_counts())
    getClusterData[i].append((next(iter((getSpeciesCluster.items())))[0],next(iter((getSpeciesCluster.items())))[1]))
```

### 2 (c)

```python
# This cell also has combine answer for 2(a), 2(b), 2(c)
hammingLoss = []
hammingDistance = []
hammingScore = []
for i in range(0,50):
    bestK = 0
    bestScore = 0
    randStateGenMC = random.randint(1,1000)
    for k in range(2,51):
        model = KMeans(n_clusters=k,random_state = randStateGenMC)
        predictions = model.fit_predict(Xdata)
        score = silhouette_score(Xdata, predictions)
        if score > bestScore:
            bestScore = score
            bestK = k
    bestmodel = KMeans(n_clusters=bestK)
    bestmodel.fit(Xdata)
    bestLabels = bestmodel.labels_
    print('---------------------------------------Iteration:', i,'-----------------------------------------------------------')
    print(f'bestK:{bestK}, silhouette score:{bestScore}')    
    itrdataLabels = []
    for k in range(0, bestK):
        getfamilyLabel = {}
        getgenusLabel = {}
        getspeciesLabel = {}
        
        for itr in range(0, len(Ydata)):
            if bestLabels[itr] == k:
                trueLabelFamily = Ydata['Family'][itr]
                if trueLabelFamily in getfamilyLabel.keys():
                    getfamilyLabel[trueLabelFamily] += 1
                else:
                    getfamilyLabel[trueLabelFamily] = 1
                
                trueLabelGenus = Ydata['Genus'][itr]
                if trueLabelGenus in getgenusLabel.keys():
                    getgenusLabel[trueLabelGenus] += 1
                else:
                    getgenusLabel[trueLabelGenus] = 1
                
                trueLabelSpecies = Ydata['Species'][itr]
                if trueLabelSpecies in getspeciesLabel.keys():
                    getspeciesLabel[trueLabelSpecies] += 1
                else:
                    getspeciesLabel[trueLabelSpecies] = 1
        
        bestKFamilyLabels = sorted(getfamilyLabel.items(), key = operator.itemgetter(1), reverse=True)[0][0]
        bestKGenusLabels = sorted(getgenusLabel.items(), key = operator.itemgetter(1), reverse=True)[0][0]
        bestKSpeciesLabels = sorted(getspeciesLabel.items(), key = operator.itemgetter(1), reverse=True)[0][0]
        dataLabels = {}
        dataLabels['Family'] = bestKFamilyLabels
        dataLabels['Genus'] = bestKGenusLabels
        dataLabels['Species'] = bestKSpeciesLabels
        itrdataLabels.append(dataLabels)
    
    familyPredictions = []
    genusPredictions = []
    speciesPredictions = []
    
    for j in range(len(Xdata)):
        getVal = bestLabels[j]
        getLabels = itrdataLabels[getVal]
        itrfamilyLabels = getLabels['Family']
        itrgenusLabels = getLabels['Genus']
        itrspeciesLabels = getLabels['Species']
        familyPredictions.append(itrfamilyLabels)
        genusPredictions.append(itrgenusLabels)
        speciesPredictions.append(itrspeciesLabels)
    
    trueFamily = np.array(Ydata['Family'])
    familyPredictions = np.array(familyPredictions)
    trueGenus = np.array(Ydata['Genus'])
    genusPredictions = np.array(genusPredictions)
    trueSpecies = np.array(Ydata['Species'])
    speciesPredictions = np.array(speciesPredictions)
    hammingL = hamming_loss(familyPredictions, trueFamily) + hamming_loss(genusPredictions, trueGenus) + hamming_loss(speciesPredictions, trueSpecies)
    hammingD = hamming_loss(familyPredictions, trueFamily) + hamming_loss(genusPredictions, trueGenus) + hamming_loss(speciesPredictions, trueSpecies)
    hammingD = (hammingD/3) * (Ydata.shape[0])
    hammingS = 1 - hammingL/3
    print(f'Hamming Loss:{hammingL/3}, Hamming Distance:{hammingD}, Hamming Score:{hammingS}')
    hammingLoss.append(hammingL/3)
    hammingDistance.append(hammingD)
    hammingScore.append(hammingS)
    for i in range(len(itrdataLabels)):
        print(f'cluster {i}:{itrdataLabels[i]}')
```

```python
from statistics import mean, pstdev
print("Average of Hamming Score", mean(hammingScore))
print("Standard deviation of Hamming Score", pstdev(hammingScore))

print("Averge of Hamming Loss", mean(hammingLoss))
print("Standard deviation of Hamming Loss", pstdev(hammingScore))

print("Average of Hamming Distance", mean(hammingDistance))
print("Standard Deviation of Hamming Distance", pstdev(hammingDistance))
```
