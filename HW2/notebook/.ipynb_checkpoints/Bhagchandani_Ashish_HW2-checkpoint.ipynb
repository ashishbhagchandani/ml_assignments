Test both models on the remaining points and report your train and test MSEs.

```python
LRresult = collections.defaultdict(list)
dataX = data[['AT','V','AP','RH']]
dataY = data[['PE']]
X_train, X_test, y_train, y_test = train_test_split(dataX, dataY, test_size=0.30, random_state=42)
```

```python
dataIntPNew = pd.concat([combineInt, dataX[['AT']]**2, dataX[['V']]**2, dataX[['AP']]**2, dataX[['RH']]**2], axis=1)
dataIntPNew.columns = ['AT','V','AP','RH','AT_V','AT_AP','AT_RH','V_AP','V_RH','AP_RH', 'pow(AT,2)','pow(V,2)','pow(AP,2)','pow(RH,2)']
```

#### (i) KNN Regression:
#### i. Perform k-nearest neighbor regression for this dataset using both normalized and raw features. Find the value of k (1 to 100) that gives you the best fit. Plot the train and test errors in terms of 1=k.

```python
KNNresults = collections.defaultdict(list)
```

```python
# raw data
dataX = data[['AT','V','AP','RH']]
dataY = data[['PE']]
X_train, X_test, y_train, y_test = train_test_split(dataX, dataY, test_size=0.30, random_state=42)
trainError = []
testError = []
ctestError = 100
indx = 0
i = 1
while i <= 100:
    knn = KNeighborsRegressor(n_neighbors = i)
    knn.fit(X_train, y_train)
    trainPredict = knn.predict(X_train)
    trainMSE = metrics.mean_squared_error(y_train, trainPredict)
    trainError.append(trainMSE)
    testPredict = knn.predict(X_test)
    testMSE = metrics.mean_squared_error(y_test, testPredict)
    testError.append(testMSE)
    if testMSE < ctestError:
        ctestError = testMSE
        indx = len(testError)
    i += 1
KNNresults['raw'] = [indx,ctestError,trainError[indx-1]]
print('K value:',indx)
print('Test Error:',ctestError)
print('Train Error:',trainError[indx-1])
```

```python
n = np.arange(1,101)
plt.rcParams.update({'figure.figsize':(8,4), 'figure.dpi':100})
plt.plot(1 / n, trainError, label="trainMSE")
plt.plot(1 / n, testError, label="testMSE")
plt.xlabel("1/k")
plt.ylabel("MSE")
plt.legend()
plt.title("Raw data - MSE versus 1/k")
plt.show()
```

```python
# normalize data
from sklearn.preprocessing import normalize
dataX = data[['AT','V','AP','RH']]
dataY = data[['PE']]
dataX = normalize(dataX)
X_train, X_test, y_train, y_test = train_test_split(dataX, dataY, test_size=0.30, random_state=42)

trainError = []
testError = []
ctestError = 100
indx = 0
i = 1
while i <= 100:
    knn = KNeighborsRegressor(n_neighbors = i)
    knn.fit(X_train, y_train)
    trainPredict = knn.predict(X_train)
    trainMSE = metrics.mean_squared_error(y_train, trainPredict)
    trainError.append(trainMSE)
    testPredict = knn.predict(X_test)
    testMSE = metrics.mean_squared_error(y_test, testPredict)
    testError.append(testMSE)
    if testMSE < ctestError:
        ctestError = testMSE
        indx = len(testError)
    i += 1
print('K value:',indx)
print('Test Error:',ctestError)
print('Train Error:',trainError[indx-1])
KNNresults['normalise'] = [indx,ctestError,trainError[indx-1]]
```

```python
n = np.arange(1,101)
plt.rcParams.update({'figure.figsize':(8,4), 'figure.dpi':100})
plt.plot(1 / n, trainError, label="trainMSE")
plt.plot(1 / n, testError, label="testMSE")
plt.xlabel("1/k")
plt.ylabel("MSE")
plt.legend()
plt.title("Normalise data - MSE versus 1/k")
plt.show()
```

## 2. ISLR 2.4.1

1. For each of parts (a) through (d), indicate whether we would generally expect the performance of a flexible statistical learning method to be better or worse than an inflexible method. Justify your answer.

(a) The sample size n is extremely large, and the number of predictors p is small.

Ans: The performance of flexible statistical learning method will be better than inflexible method for the most of the cases as due to large number of sample size n it will reduce the option of overfitting and with number of predictors p being small their will be less Redundancy, Overfitting, and Productivity.

(b) The number of predictors p is extremely large, and the number of observations n is small.

Ans: The performance of inflexible statistical learning method will be better than flexible method. As flexible method has high variance therefore its model will change a lot also with number of predictors p being extremely high flexible statistical learning method will tend to overfit.

(c) The relationship between the predictors and response is highly non-linear.

Ans: Inflexible method does not performs well when it comes to non-linear relations whereas flexible method performs well on non-linear relations, has a better model fit and approximate data predicted with less error.

(d) The variance of the error terms, i.e. Ïƒ2 = Var(Ïµ), is extremely high.

Ans: As variance of the error terms is extremely high this means the error terms will have high noise. Therefore on this inflexible method will perform well as it will have less chance of overfitting compare to flexible method.
